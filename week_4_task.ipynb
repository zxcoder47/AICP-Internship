### Q.3: Check to see if infact there is any correlation between Calories and other independent variables by plotting a correlation matrix next.

```python
# Select the relevant columns for the correlation matrix
columns_of_interest = [
    'Calories',
    'Total Fat',
    'Saturated Fat',
    'Cholesterol',
    'Sodium',
    'Carbohydrates',
    'Dietary Fiber',
    'Sugars',
    'Protein',
    'Vitamin A (% Daily Value)',
    'Vitamin C (% Daily Value)',
    'Calcium (% Daily Value)',
    'Iron (% Daily Value)'
]

selected_columns = menu_df[columns_of_interest]

# Calculate the correlation matrix
correlation_matrix = selected_columns.corr()

# Plot the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

### Q.4: Draw boxplot for Calories vs Category to spot outliers and max calories category

### Q.7: Draw a horizontal bar graph for items in each category against the calories. 
Here is one sample for all items in Beef & Pork, you have to do it for items in each category. Also, write
 your observation.

```python
# Select the relevant columns
category_calories_df = menu_df[['Category', 'Item', 'Calories']]

# Group the data by category and calculate the average calories for each item in the category
category_avg_calories_df = category_calories_df.groupby(['Category', 'Item']).mean().reset_index()

# Sort the data by calories in descending order within each category
category_avg_calories_df.sort_values(['Category', 'Calories'], ascending=[True, False], inplace=True)

# Set up the figure and axes
plt.figure(figsize=(10, 8))
ax = sns.barplot(x='Calories', y='Item', hue='Category', data=category_avg_calories_df, dodge=False)

# Set the x-axis label
plt.xlabel('Calories')

# Set the y-axis label
plt.ylabel('Item')

# Set the title
plt.title('Average Calories of Items in Each Category')

# Show the legend
plt.legend(title='Category')

# Show the plot
plt.show()
```

**Observation:** By examining the horizontal bar graph, you can observe the relative average calorie content of items within each category. The graph allows for easy visual comparison of the calorie levels across different items in the same category. You can identify the items with higher or lower average calories within each category based on the length of the bars.
